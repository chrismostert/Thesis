```python
from pathlib import Path
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import math
import numpy as np
# Use seaborn style defaults and set the default figure size
sns.set(rc={'figure.figsize':(8, 8)})

from tqdm.notebook import tqdm
tqdm().pandas()


# Load in the acousticbrainz dataset into the variable 'acousticbrainz'
acousticbrainz = pd.read_hdf(Path.cwd() / 'datasets' / 'acousticbrainzV3.h5')

# Explicitly add mbid as non-index to make intersection easier
acousticbrainz['mbid'] = acousticbrainz.index.get_level_values(0)

# Classifiers to select
toSelect = [('mood_acoustic', 'acoustic'), ('mood_electronic', 'electronic'),
            ('mood_relaxed', 'relaxed'), ('mood_sad', 'sad')]
# Labels used for these plots
labels = [x[0] for x in toSelect]

# Spotify classifiers to select
toSelect_spotify = ['acousticness', 'danceability', 'energy', 'instrumentalness', 'valence']
# Labels used for these plots
labels_spotify = toSelect_spotify

def plot_distribution(data, toSelect, labels, ytop):
    cols = [data[x] for x in toSelect]
    sns.set(font_scale=1)
    fig = plt.figure(figsize=(8,8))
    plt.ylim(top=ytop)
    plt.xlabel('Bin range')
    plt.ylabel('Bin count')

    for i in range(0, len(cols)):
        y,binEdges = np.histogram(cols[i],bins=200)
        bincenters = 0.5*(binEdges[1:]+binEdges[:-1])
        plt.plot(bincenters, y, label=labels[i])

    plt.legend()
    plt.show()
```

# Full dataset (all.txt)

```python
# Make the plot
plot_distribution(acousticbrainz, toSelect, labels, 500000)

# Size stats
print(f"Amount of recordings (MBIDs): {len(list(acousticbrainz.index.get_level_values(0).unique()))}")
print(f"Amount of submissions: {len(acousticbrainz)}")

# Dump the MBIDs
with open(Path.cwd() / 'mbid dumps' / 'all.txt', 'w') as f:
    f.write('\n'.join(list(acousticbrainz.index.get_level_values(0).unique())))
    
```

# Filtered on at least 2 submissions (filtered_multiple_submissions.txt)

```python
filt = acousticbrainz.groupby(level=0).size() > 1
acousticbrainz_filt = acousticbrainz[filt[acousticbrainz.index.get_level_values(level=0)].values]

# Make the plot
plot_distribution(acousticbrainz_filt, toSelect, labels, 270000)

# Size stats
print(f"Amount of recordings (MBIDs): {len(list(acousticbrainz_filt.index.get_level_values(0).unique()))}")
print(f"Amount of submissions: {len(acousticbrainz_filt)}")

# Dump the MBIDs
with open(Path.cwd() / 'mbid dumps' / 'filtered_multiple_submissions.txt', 'w') as f:
    f.write('\n'.join(list(acousticbrainz_filt.index.get_level_values(0).unique())))
```

# Filtered on artist == 'The Beatles' (filtered_beatles_songs.txt)

```python
beatles = pd.read_hdf(Path.cwd() / 'datasets' / 'beatles.h5')

# Make the plot
plot_distribution(beatles, toSelect, labels, 3500)

# Size stats
print(f"Amount of recordings (MBIDs): {len(list(beatles.index.get_level_values(0).unique()))}")
print(f"Amount of submissions: {len(beatles)}")

# Dump the MBIDs
with open(Path.cwd() / 'mbid dumps' / 'filtered_beatles_songs.txt', 'w') as f:
    f.write('\n'.join(list(beatles.index.get_level_values(0).unique())))
```

# Intersected with spotify via MSD mapping (spotify_msd_intersection.txt)

```python
spotify = pd.read_hdf(Path.cwd() / 'datasets' / 'spotify.h5')
mapping = pd.read_hdf(Path.cwd() / 'datasets' / 'mapping.h5')
acousticbrainz_spotify = acousticbrainz.merge(mapping, on='mbid').merge(spotify, left_on='spotify', right_on='id')

# Make the plot for acousticbrainz data
print("Acousticbrainz side (can contain multiple submissions per MBID):")
plot_distribution(acousticbrainz_spotify, toSelect, labels, 80000)

# Size stats
print(f"Amount of recordings (MBIDs): {len(list(acousticbrainz_spotify['mbid'].unique()))}")
print(f"Amount of submissions: {len(acousticbrainz_spotify)}")

print()

# Make the plot for spotify data
print("Spotify side (contains only one submission per Spotify ID):")
plot_distribution(spotify, toSelect_spotify, labels_spotify, 4000)

# Size stats
print(f"Amount of recordings (spotify IDs): {len(list(spotify['id'].unique()))}")

# Dump the MBID, spotifyID pairs
mapping[['mbid', 'spotify']].sort_values('mbid').to_csv(Path.cwd() / 'mbid dumps' / 'spotify_msd_intersection.txt',
                                                        index=False)
```

# Intersected with AcousticBrainz genre dataset (genre_dataset_intersection.txt)

```python
# We have three 'ground truth sources' for genre: discogs, lastfm and tagtraum
discogs = pd.read_table(Path.cwd() / 'datasets' / 'acousticbrainz_genres' / 'acousticbrainz-mediaeval2017-discogs-train.tsv')
lastfm = pd.read_table(Path.cwd() / 'datasets' / 'acousticbrainz_genres' / 'acousticbrainz-mediaeval2017-lastfm-train.tsv')
tagtraum = pd.read_table(Path.cwd() / 'datasets' / 'acousticbrainz_genres' / 'acousticbrainz-mediaeval2017-tagtraum-train.tsv')

# Check which MBIDs are contained in all three datasets
discogs_ids = set(discogs['recordingmbid'].unique())
lastfm_ids = set(lastfm['recordingmbid'].unique())
tagtraum_ids = set(tagtraum['recordingmbid'].unique())
genre_ids = discogs_ids.intersection(lastfm_ids, tagtraum_ids)

genre_filtered = acousticbrainz.loc[list(genre_ids)]
```

```python
# Make the plot
plot_distribution(genre_filtered, toSelect, labels, 125000)

# Size stats
print(f"Amount of recordings (MBIDs): {len(list(genre_filtered.index.get_level_values(0).unique()))}")
print(f"Amount of submissions: {len(genre_filtered)}")

# Dump the MBIDs
with open(Path.cwd() / 'mbid dumps' / 'genre_dataset_intersection.txt', 'w') as f:
    f.write('\n'.join(list(genre_ids)))
```
